```python
import matplotlib.pyplot as plt
import pandas as pd

# Data from the user's table
data = {
    "Models": ["ChronoPatternNet", "TCN", "LSTM", "GRU", "MLP"],
    "Number of Parameters": [61, 148, 151, 148, 160],
    "MSE": [0.0055, 0.0075, 0.0061, 0.007, 0.0065]}

# Creating a DataFrame
df = pd.DataFrame(data)
# Adjusting the plot with high definition and larger size


# Creating the plot
fig, ax1 = plt.subplots(figsize=(10, 8))

# Plotting Number of Parameters
color = 'tab:blue'
# ax1.set_xlabel('Models')
ax1.set_ylabel('Number of Parameters', color=color)
ax1.bar(df['Models'], df['Number of Parameters'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Plotting MSE on a second y-axis
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('MSE', color=color)
ax2.plot(df['Models'], df['MSE'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Adjusting the layout for clarity
fig.tight_layout()
plt.title('Comparison of Number of Parameters and MSE for Various Models')
plt.show()

```
